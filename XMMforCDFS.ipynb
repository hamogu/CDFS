# XMM-Newton observations of the Chandra deep field South

There are 47 XMM-Newton observations of the regions best know as the Chandra Deep Field South (CDFS). The official names of the field in the observations is either "AXAF Ultra Deep F" or later "CDFS". They can be found in the XMM-Newton Science Archive (XSA) with a seach on position for RA=53.1488 and DE=-27.7554. The separate observations have very different exposure times between 3 and 130 ks. In one case (ObsID 0108061601) no valid data was taken and thus no pipeline processing exists. The data is taken in six blocks in 2001, 2002, 2008, 2009 and 2010 (see plot below).

In this file I want to keep my notes on how I did the processing, what I did and what we get (and also journal the ideas of what else I can try). My first approach will be to download the pipeline data products (PPS) and see how far we can get with those. Using the existing 18 GB of pipeline products is much less work than crunching through the data extraxtion of 46 obsIDs (zipped ODFs take about 10 GB) all with many different exposures myself. However, the PPS products are extracted only for souces that are highly significant in a source dection in an individual exposure. In the case of teh CDFS we have the source positions of all stars and we can probably take that data and combine it. In most blocks there are 10 or so ObsIDs with 3 EPIC detectors each, so we can go significantly deeper than in individual exposures (not 30 times deeper, because the exposures have very different exposures times. If I use the long individual exposures of 130 ks, I will only gain a factor of 3 or 4 by adding all up, because many exposures are short and don't have many counts).
The drawback is that there really is no perfect way to combine XMM data from different times and different detectors. The recommended method is to extract each detector and each exposure separately and then fit them simultaneously. However, for all but one of the stars we will only get very few counts and we will certainly have a hard time in XSPEC or Sherpa to fit 30 spectra at the same time. Maybe we skip the short exposures and only analyze the long ones? Alternatively, I could combine the spectra using the SAS tasks _multiespecget_ and _epicspeccombine_. They should preserve the statistical properties of the spectra reasonably well, but not perfect.

## Using the PPS products

```python
epicsourcelists = glob(datadir + '*/pps/P*EPX000OBSMLI0000.FTZ')
```

The PPS has one source list per observation, but our sources of interest may or may not be in there (because they might or might not be detected in this observation). So, I need to crossmatch the that source list with the positions of the stars. The position in the EPIC source list may differ from observation to observation because it is measured fro mthe observed data without any prior knowledge of the field. The *Chandra* sources may also be so close to each other that the match with XMM is not unique. I need to watch out for those cases, but as long as they don't involve anything that we believe to be a star I don't care.

```python
def find_stars_in_ppssourcelist(cdfs_starlist, cdfs_alllist, sources, note=''):
    idsall = makecrossids(cdfs_alllist, sources, sources['RADEC_ERR']/3600.*3, ra1='RA', dec1='DEC', ra2='RA', dec2='DEC')
    idstars = makecrossids(cdfs_starlist, sources, sources['RADEC_ERR']/3600.*3, ra1='RA', dec1='Dec', ra2='RA', dec2='DEC')
    for i in set(idstars[idstars >= 0]):
        if (idsall==i).sum() > 1:
            print note, 'double match to star - investigate!'
    return idstars
```

```python
results = [None] * len(cdfs_starlist)

for f in epicsourcelists:
    sources = Table.read(f)
    sources.add_column(Column(name='ObsID', data=[fits.getval(f, 'OBS_ID')]*len(sources)))
    sources.add_column(Column(name='DATE-OBS', data=[fits.getval(f, 'DATE-OBS')]*len(sources)))
    idstars = find_stars_in_ppssourcelist(cdfs_starlist, cdfs_alllist, sources, note=sources['ObsID'][0])
    for i, ids in enumerate(idstars):
        if ids >=0:
            if results[i] is None:
                results[i] = sources[[ids]]
            else:
                # in the current version astropy will issue and error, if the same key appears in both headers. So, empty out the header of the new row before merging.
                # (This is an known astropy problem, waiting to be resolved.)
                newrow = sources[[ids]]
                newrow.meta.clear()
                results[i] = astropy.table.vstack([results[i], newrow])
```
